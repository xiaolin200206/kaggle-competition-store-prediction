```python
#handing missing value
print(train.isna().sum())

# what's ffill and bfill??
# ffill mean use previously value to fill missing value
# bfill mean use back values to fill missing value
train.fillna(method="ffill", inplace = True) #ffill means forward fill 
train.fillna(method="bfill", inplace = True) #bfill means backward fill 
```

```python
train["date"] = pd.to_datetime(train["date"])
train.set_index("date", inplace=True)
```

```python
#feature enginerring
#feature enginerring is  a process of extracting useful feature from raw data using math, statistic and
#domain knowledge
#why need to index here?? time series model like lstm rely on the order of data
#allow easy access to data-based attributed
#make resampling,rolling statistics and lag feature easier
train["date"] = train.index.date
train["month"] = train.index.month
train["day"] = train.index.day
train["dayofweek"] = train.index.dayofweek
train["is_weekend"] = (train["dayofweek"] >= 5).astype(int)  #why set is_week >= 5? because 5 equal saturday

#lag feature
#why are lag feature important?? lag feature help lightgbm learn past sale patterns
# and can detect trends
# this model can help capture seasonality such as weekly,monthly partterns
train["lag_feature"] = train["sales"].shift(7) #capture weekly trend
train["lag_feature"] = train["sales"].shift(14) #help with bi-weekly trend

#why use rolling statistic? can help us trend, anomaly detection and feature engineering
#rolling statistics for improves trend detection
#rolling statistic used to smooth out time series data and capture trends over time
train["rolling_mean_7"] = train["sales"].rolling(window = 7).mean() #this mean take the average of the previous
#7 days sale

#rolling standard deviation measure how much the values vary within a moving window 
train["rolling_std_7"] = train["sales"].rolling(window = 14).std()
#this code helps to detect sudden spike or drop in sales
```

```python
#must we use resampling, rolling, and lag when we use lstm model??
#no, not require, but they are very useful in different scenarios
# when will we use resample? if you time series data has missing timestamp
# if your data has a high frequency and you want convert it to daily
#when you dataset is already clean and has a consistent time step is not needed

#when will we use rolling?? help smooth out fluctuations and capture long-term trends
#can serve as a feature for LSTM, help the model understand past behavior
#when you model is powerful enough to learn trends without manual smoothing is not needed

#when will we use lag?? lag is extremely useful for lstm, because they provide past sale data explicitly
#LSTM already own memoru, add lag can help it learn dependence more faster
#if you dataset is small, add too much lag will reduce training samples
```

```python
#scaling data
#what's scaler?
#scaler is a tool used in machine learning to normalize or standardize your data feature, ensure that they are on a similar scale
#this can help many algorithms perform better, because they often work best when the input feature are similarly scaled
#for example, the standardscaler from scikit-learning standardizes features by removing the mean and scaling to unit variance
#why?? when feature are on different scales, some models can be biased toward feature with large ranges
#standardizing help algorithms converge faster and perform better
#scaler.fit(x) calculate the mean and standard deviation for each feature in your dataset x
#the scaler learns from x: 
     #mean: average value of each feature
     #std: measure of how spread out the feature value are

#scaler.transform(x): applies the transformation to your data using the previously computed mean and std
#each feature in x is standardized:
     #the transformed feature will have a mean of 0
    #the transformed feature will have a standard deviation of 1

#x = scaler.fit_transform(X): this method fit the scaler to x(computing mean and std) and then transforms x in one go

# when to use it??
#before training: we will use it to scaler your training data, and then use the same scaler to transform your test data. this ensure that the data
#has the same distribution

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
train["sales"] = scaler.fit_transform(train[["sales"]]) #transform: both fit(calculates statistics) and transform(applies scaling)
#[["sales"]] ensure input is a 2d array dataframe
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train["family"] = le.fit_transform(train["family"])
train["is_holiday"] = le.fit_transform(train["is_holiday"])
train["date"] = le.fit_transform(train["date"])

test["family"] = le.fit_transform(test["family"])
test["is_holiday"] = le.fit_transform(test["is_holiday"])
test["date"] = le.fit_transform(test["date"])
```

```python
model.fit(train.drop(columns=["sales"]), train["sales"])
#drop(columns=["sale"] is input, used to predict
#train["sales"] is target, indicate what we are predicting 
#why do we need to drop"sales"??? because it's convenient for model only learns from feature
# and doesn't cheat
#why we keep sales as y?? it can let model knows what to predict
# this is a standard process in machine learning to prevent data leakage 
```

```python
from sklearn.metrics import root_mean_squared_error
#make prediction & evaluate performance
#prediction using the trained model to train, why??
#trainin is very expensive and waste time
#maybe can cause overfitting 
#the model already ready for prediction since you trained, like you don't need to study every time
#how to understan this output??
#rmse measure how far your predicted sales are from actual sales
#if value is low mean better model accuracy, if vallue is higher mean larger prediction errors
pred = model.predict(test.drop(columns=["sales"]))
print("RMSE:",root_mean_squared_error(test["sales"], pred))
```

```python
from sklearn.model_selection import GridSearchCV 
# "num_leave" : [31,50] mean test two values for num_leaves parameter 31 and 50
# "learning_rate": [0.01,0.1] mean  test two values for the learning_rate parameter:0.1 and 0.01
#what's this for?? this create different combination of parameters to find the best-performing one
param_grid = {"num_leaves" : [31,50], "learning_rate": [0.01,0.1]}
#LGBMRegreeor automatically tests all combinations of parameters
#cv = 3 mean use 3-fold cross-validation, mean the dataset is split into 3 parts
#and the model is trained on different split to find the best parameter
grid = GridSearchCV(lgb.LGBMRegressor(), param_grid, cv=3)
# train.drop(columns=["sales"] use all features except the target(sales)
# train["sales"] is the target variable
grid.fit(train.drop(columns=["sales"]), train["sales"])
print(grid.best_params_)
```

```python
#what's parameters,hyperparameter and hyperparameter tune mean 
#parameter in a mechine learning model are the parameter whose values are updated during training using some optimization procedure

#hyperparamter in a machine learning model are the parameters whose values are decided before training of the model begin.
#hyperparameter are not affect(do not change)by training, rather, these affect the quality and speed of the training.
#so in a sense, these parameters are over, beyond or above the process of training and hence called hyperparameter
#and hyperparameter can be related to model selection for example:model type and learning rate

#there are no efficient algorithms to select optimal values of hyperparameters. so optimal values of hyperparameters are determined
#using a trial and error process by adjusting or adapting their value for a particular machine learning task. this process of determined
#values of hyperparameters is called hyperparameter tuning

#why need to hyperparameter tuning
#hyperparameter tuning is essential because different value of hyperparameter significantly effect a 
#model's performance. if we don't tune hyperparameter, the model might:
#1.very simple, and poor performance
#2.very complex, it'll memorizes training data cause bad generalization
#3.very slow and high computation cost

#what if our don't tune hyperparameter??
#default value may work, but they are not optimized for your dataset
#the model will poor genelization if the perameters are not well-fitted 
#will spend too much time in training time(too many trees in a random forest)
#the model accuracy will low because of suboptimal settings

#different between parameter and hyperparameter
#let we understand they definition first 
#parameter: this is value learned during training from the dateset and can automatically adjusted
#hyperparameter: it set before training and must manually or using GridSearchCV,RandomizedSearchCV
# parameter changes in per dataset, whereas hyperparameter fix before training

#when do we use hyperparameter??
#control model complexity, optimize training speed, improve accuracy, balance bias and variance
#handle large dataset 

#when do we use parameter??
#after training we use parameter to make prediction and use parameter to evaluating model performance,
#model's accuracy depend on well-tune parameter. if parameter is not good then we will change 
#hyperparameter and retrain, in some model like linear regression, parameter can tell us which features
#matter cost
```
